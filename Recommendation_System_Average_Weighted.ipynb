# Average Weighted Recommendation System

Consider a new user came created account and we have to provide suggest best movies from our list. Here, we have no records of user choices so it is hard to suggest movie to the user. Instead we create a general recommendation system i.e. the movie that is popular among 70% of the viewers and also have high rating.

This technique of recommendation system is Weighted Average technique. Here, we consider the movie rating by all users (or mean rating), the number of votes movie obtain, mean value of all the ratings of all the movies and 70% of count.

Here, we are using dataset from Kaggle competetion which can be downloaded from 

https://www.kaggle.com/tmdb/tmdb-movie-metadata

### Importing libraries

```python
import numpy as np
import pandas as pd
```

### Loading data and Data Analysis

```python
movies = pd.read_csv("TMDB-dataset/tmdb_5000_movies.csv")
movies.head()
```

```python
credits = pd.read_csv("TMDB-dataset/tmdb_5000_credits.csv")
credits.head()
```

```python
credits.rename(columns={'movie_id':'id'}, inplace=True)
credits.head()
```

```python
df = movies.merge(credits, on='id')
df.head()
```

```python
df.info()
```

```python
columns_to_be_dropped = ["budget", "homepage", "title_x", "title_y", "cast", "crew", "production_companies",
                         "original_language", "release_date", "revenue", "spoken_languages", "status"
                        ]
df.drop(columns=columns_to_be_dropped, inplace=True)
df.head()
```

### Formula for Weighted Average

For Weighted Average: 

                W = (Rv + cm)/(v+m)
                
                where:
                W = Weighted Rating
                R = average for movie rating
                v = number of votes
                m = minimum votes required 
                c = the mean value across the all the movies

```python
R = df['vote_average']
v = df['vote_count']
c = df.vote_average.mean()
m = df.vote_count.quantile(0.7)
```

```python
df['Weights'] = (R*v + c*m)/(v+m)
df.head()
```

## Sorting The Movie Based On Scores

```python
movie_order = df.sort_values(by=['Weights'], ascending=False)
movie_order[['original_title', 'vote_count', 'vote_average', 'Weights', 'popularity']].head(20)
```
