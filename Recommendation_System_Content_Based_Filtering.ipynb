# Recommender System Content Based Filtering

When a user is listening to a song of any genre, he is more likely to listen the song of same genre. E.g. a person listening to hip-hop is not likely to move to the sad songs or on who is listening to english may not suddenly shift to korean or irish songs. User would be willing to continue to same trend.

This is how Content Based Filtering works. For this type of recommendation system, we take either one or different attributes of item. E.g. for movies, language, country of origin, generes, series etc.

### Importing libraries

```python
import numpy as np
import pandas as pd
```

### loading data

```python
credits = pd.read_csv("TMDB-dataset/tmdb_5000_credits.csv")
movies = pd.read_csv("TMDB-dataset/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
credits.rename(columns={'movie_id':'id'}, inplace=True)
credits.head()
```

### Data Preparation

```python
df = movies.merge(credits, on='id')
df.head()
```

```python
df.info()
```

Here, We are going for recommendation based on the overview i.e. the story of the movie. So, we can merge the data and than transform the overview sentences into numbers.

```python
df.overview.isna().sum()
```

```python
df.overview = df.overview.fillna(' ')
df.overview.isna().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer()
```

```python
overview_matrix = tfv.fit_transform(df['overview'])
overview_matrix
```

### Finding Similarity Score Based on Overview

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(overview_matrix)
```

```python
sig[0]
```

```python
# Reverse mapping if indices and titles
indices = pd.Series(df.index, index=df['original_title']).drop_duplicates()
indices
```

### Sorting Movie with Highest Similarity

```python
movie_name = "Newlyweds"
```

```python
sorted(list(enumerate(sig[indices[movie_name]])), key=lambda x: x[1], reverse=True)
```

### Recommendation Function

```python
def movie_recommend(title, sig=sig):
    idx = indices[title]
    
    sig_scores = list(enumerate(sig[idx]))
    
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
    sig_score = sig_scores[1:11]
    
    movie_indices = [i[0] for i in sig_score]
    
    return df['original_title'].iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the 'Avatar'
movie_recommend('The Godfather')
```
